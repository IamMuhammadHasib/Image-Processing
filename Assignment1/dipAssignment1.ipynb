```python
import matplotlib.pyplot as plt
import cv2


def main():
    img_path = './image.jpg'
    image = plt.imread(img_path)
#     print(image.shape)
#     print(image)
    
    plt.figure(figsize = (30, 60))
    
    plt.subplot(6,2,1)
    plt.title('Main Image')
    plt.imshow(image)
    
    
    plt.subplot(6,2,3)
    plt.title('Red Channel')
    red_ch = image[:, :, 0]
    plt.imshow(red_ch, cmap = 'Reds')
    
    plt.subplot(6,2,4)
    plt.title('Red Channel Histogram')
    plt.hist(red_ch.ravel(), 256, [0,256])
    
    
    plt.subplot(6,2,5)
    plt.title('Green Channel')
    green_ch = image[:, :, 1]
    plt.imshow(green_ch, cmap = 'Greens')
    
    plt.subplot(6,2,6)
    plt.title('Green Channel Histogram')
    plt.hist(green_ch.ravel(), 256, [0,256])
    
    
    plt.subplot(6,2,7)
    plt.title('Blue Channel')
    blue_ch = image[:, :, 2]
    plt.imshow(blue_ch, cmap = 'Blues')
    
    plt.subplot(6,2,8)
    plt.title('Blue Channel Histogram')
    plt.hist(blue_ch.ravel(), 256, [0,256])

    
    plt.subplot(6,2,9)
    plt.title('Grayscale Image')
    grayscale_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    plt.imshow(grayscale_img, cmap = 'gray')
    
    plt.subplot(6,2,10)
    plt.title('Grayscale Image Histogram')
    plt.hist(grayscale_img.ravel(), 256, [0,256])
    
    
    plt.subplot(6,2,11)
    plt.title('Binary Image')
    th, binary_img = cv2.threshold(grayscale_img, 127, 256, cv2.THRESH_BINARY)
    plt.imshow(binary_img, cmap = 'binary')
    
    plt.subplot(6,2,12)
    plt.title('Binary Image Histogram')
    plt.hist(binary_img.ravel(), 256, [0,256])
    
    plt.savefig('processed.png')
    
if __name__ == '__main__':
    main()
```
