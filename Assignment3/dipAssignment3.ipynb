```python
import matplotlib.pyplot as plt
import cv2
import math
import numpy as np

img_path = './image.jpg'
image = plt.imread(img_path)
# print(image)

T1 = 99
T2 = 201
c = 25
p = 3
epsilon = 0.0000001

plt.figure(figsize = (30, 30))

plt.subplot(4,2,1)
plt.title('Main Image', fontsize = 30)
plt.imshow(image)

plt.subplot(4,2,2)
plt.title('Grayscale Image', fontsize = 30)
grayscale_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(grayscale_img, cmap = 'gray')

# print(grayscale_img.shape)
# print(grayscale_img)

width = grayscale_img.shape[1]
height = grayscale_img.shape[0]



kernel1 = np.array([[-1,-1,-1], [-1,8,-1], [-1,-1,-1]])
grayscale_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
c1_img = cv2.filter2D(grayscale_img, -1, kernel1)
print(kernel1)
print()

kernel2 = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])
grayscale_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
c2_img = cv2.filter2D(grayscale_img, -1, kernel2)
print(kernel2)
print()

kernel3 = np.array([[-2,-1,0], [-1,1,1], [0,1,2]])
grayscale_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
c3_img = cv2.filter2D(grayscale_img, -1, kernel3)
print(kernel3)
print()

kernel4 = np.array([[1,0,-1], [2,0,-2], [0,1,-1]])
grayscale_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
c4_img = cv2.filter2D(grayscale_img, -1, kernel4)
print(kernel4)
print()

kernel5 = np.array([[0.008,0.008,0.008], [0.008,0.008,0.008], [0.008,0.008,0.008]])
grayscale_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
c5_img = cv2.filter2D(grayscale_img, -1, kernel5)
print(kernel5)
print()

kernel6 = np.array([[0,-8,0], [-8,40,-8], [0,-8,0]])
grayscale_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
c6_img = cv2.filter2D(grayscale_img, -1, kernel6)
print(kernel6)
print()

plt.subplot(4,2,3)
plt.title('Karnel 1', fontsize = 30)
plt.imshow(c1_img, cmap = 'gray')

plt.subplot(4,2,4)
plt.title('Kernel 2', fontsize = 30)
plt.imshow(c2_img, cmap = 'gray')

plt.subplot(4,2,5)
plt.title('Kernel 3', fontsize = 30)
plt.imshow(c3_img, cmap = 'gray')

plt.subplot(4,2,6)
plt.title('Kernel 4', fontsize = 30)
plt.imshow(c4_img, cmap = 'gray')

plt.subplot(4,2,7)
plt.title('Kernel 5', fontsize = 30)
plt.imshow(c5_img, cmap = 'gray')

plt.subplot(4,2,8)
plt.title('Kernel 6', fontsize = 30)
plt.imshow(c6_img, cmap = 'gray')

plt.savefig('processed-image.png')
# plt.show()
```
